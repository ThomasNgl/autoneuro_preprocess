```python
import pandas as pd
import os 
path_to_lafolette_exp1 = '/Users/thomasnegrello/Desktop/autoneuro/datasets/lafolette_2023/'
df = pd.read_csv(path_to_lafolette_exp1+'study2_processed.csv')
# Remove practice
df = df[df['trial_type']!='preattentioncheck']
df = df[df['trial_type']!='attentioncheck']
```

```python
df
```

```python
id_col = df['ID']
state_col = df['working_phone']

reward_col = df['working_phone']
value_col = df['reported_q']

reward_col = reward_col.shift(-1)
reward_col.iloc[-1] = -1  

value_col = value_col.shift(-1)
value_col.iloc[-1] = -1  

# Create new DataFrame using the generated columns
fdf = pd.DataFrame({
    'id': id_col,
    's': state_col,
    'v': value_col,
    'r': reward_col
})
```

```python
nb_subjects = int(fdf['id'].iloc[-1])
path_to_processed_lafolette_exp1 = '/Users/thomasnegrello/Desktop/autoneuro/datasets/datasets_processed/dataset_lafolette_exp2/'

# Save the required columns to their respective files
column_to_filename = {
    's': 'observations.csv',
    'v': 'values.csv',
    'r': 'rewards.csv',
}
```

```python
for id in range(1, nb_subjects+1):
    subject_id =  int(id)
    subject = fdf[fdf['id'] == subject_id].reset_index()
    # Create the directory for the subject
    subject_folder = os.path.join(path_to_processed_lafolette_exp1, 'Sub' + str(subject_id))
    os.mkdir(subject_folder)

    for col, filename in column_to_filename.items():
        file_path = os.path.join(subject_folder, filename)
        subject[col].to_csv(file_path, header=False, index=False)
```
